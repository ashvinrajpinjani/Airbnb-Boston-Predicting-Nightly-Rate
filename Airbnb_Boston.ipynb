```python
import pandas as pd

# Assuming 'subset_data' is your DataFrame

# Step 1: Create a mapping from Zipcode to Neighborhood
zipcode_to_neighborhood = subset_data.dropna(subset=['Neighborhood', 'Zipcode'])
zipcode_to_neighborhood = zipcode_to_neighborhood.groupby('Zipcode')['Neighborhood'].agg(pd.Series.mode)

# Step 2: Fill missing Neighborhood values based on Zipcode
def fill_neighborhood(row):
    if pd.isna(row['Neighborhood']) and row['Zipcode'] in zipcode_to_neighborhood:
        return zipcode_to_neighborhood[row['Zipcode']]
    else:
        return row['Neighborhood']

subset_data['Neighborhood'] = subset_data.apply(fill_neighborhood, axis=1)

# Check results
print(subset_data[['Zipcode', 'Neighborhood']].head())
```

```python
import xgboost as xgb
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
import numpy as np

# Define the MAPE function
def mean_absolute_percentage_error(y_true, y_pred):
    """Calculate the mean absolute percentage error from y_true and y_pred"""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # Avoid division by zero
    y_true = np.where(y_true == 0, np.nan, y_true)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return np.nanmean(mape)

# Predicting on the test set
y_pred = model_selected.predict(X_test)

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")


# Calculate MAPE
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")


# Optional: Feature Importance
print("\nFeature Importance in the Selected Model:")
for idx in sorted_idx_selected:
    print(f"{X_train.columns[idx]}: {feature_importance_selected[idx]}")
```

```python
# # List of initial features for backward selection
# initial_features = [
#     'ListingType_Entire home/apt', 'Bedrooms', 'Bathrooms', 'Cleaning Fee (USD)',
#     'SuperhostRatio_Zipcode', 'numReviews_pastYear', 'prev_numReserv_pastYear',
#     'Max Guests', 'Neighborhood_South End', 'Superhost_Status_Frequency',
#     'Bathroom_Bedroom_Ratio', 'PropertyType_House', 'Neighborhood_Fenway/Kenmore',
#     'Number of Photos', 'num_5_star_Rev_pastYear', 'rating_ave_pastYear',
#     'ListingsCount_Zipcode', 'Pets Allowed', 'available_days',
#     'PropertyType_Serviced apartment', 'PropertyType_Entire apartment',
#     'Neighborhood_Downtown', 'PropertyType_Apartment', 'Neighborhood_Beacon Hill',
#     'PropertyType_Room in boutique hotel', 'PropertyType_Townhouse',
#     'PropertyType_Condominium', 'Instantbook Enabled', 'Neighborhood_West End',
#     'OccupancyRate_Zipcode'
#     # Add other interaction terms as needed
#]
##################################


import numpy as np
import pandas as pd
import statsmodels.api as sm
from itertools import combinations

# Function to handle missing values
def handle_missing_values(df):
    return df.fillna(df.median())

# Function to create interaction terms
def create_interaction_terms(df, feature_list):
    df = df.copy()
    for combo in combinations(feature_list, 2):
        interaction_term = f"{combo[0]}*{combo[1]}"
        df[interaction_term] = df[combo[0]] * df[combo[1]]
    return df

# Function for backward selection
def backward_selection(data, target, features):
    remaining_features = features.copy()
    best_p = 1
    while len(remaining_features) > 0:
        features_with_constant = sm.add_constant(data[remaining_features])
        p_values = sm.OLS(data[target], features_with_constant).fit().pvalues[1:] # Excluding the constant
        max_p_value = p_values.max()
        if max_p_value > 0.05: # Threshold for p-value
            excluded_feature = p_values.idxmax()
            remaining_features.remove(excluded_feature)
        else:
            break
    model = sm.OLS(data[target], sm.add_constant(data[remaining_features])).fit()
    return model, remaining_features

# Assuming subset_data is your DataFrame
features = [
    'ListingType_Entire home/apt', 'Bedrooms', 'Bathrooms', 'Cleaning Fee (USD)',
    'SuperhostRatio_Zipcode', 'numReviews_pastYear', 'prev_numReserv_pastYear',
    'Max Guests', 'Neighborhood_South End', 'Superhost_Status_Frequency',
    'Bathroom_Bedroom_Ratio', 'PropertyType_House', 'Neighborhood_Fenway/Kenmore',
    'Number of Photos', 'num_5_star_Rev_pastYear', 'rating_ave_pastYear',
    'ListingsCount_Zipcode', 'Pets Allowed', 'available_days',
    'PropertyType_Serviced apartment', 'PropertyType_Entire apartment',
    'Neighborhood_Downtown', 'PropertyType_Apartment', 'Neighborhood_Beacon Hill',
    'PropertyType_Room in boutique hotel', 'PropertyType_Townhouse',
    'PropertyType_Condominium', 'Instantbook Enabled', 'Neighborhood_West End',
    'OccupancyRate_Zipcode'
    # ...
]

# Handle missing values
subset_data_clean = handle_missing_values(subset_data[features + ['Nightly Rate']])

# Create interaction terms
subset_data_with_interactions = create_interaction_terms(subset_data_clean, features)

# Convert all columns to numeric, errors='coerce' will replace non-convertible values with NaN
subset_data_with_interactions = subset_data_with_interactions.apply(pd.to_numeric, errors='coerce')

# Replace inf/-inf with NaN and then drop rows with NaN values
subset_data_with_interactions.replace([np.inf, -np.inf], np.nan, inplace=True)
subset_data_with_interactions.dropna(inplace=True)

# Run the backward selection
final_model, final_features = backward_selection(subset_data_with_interactions, 'Nightly Rate', features)

# Print final model summary and selected features
print(final_model.summary())
print("Selected features:", final_features)
```
